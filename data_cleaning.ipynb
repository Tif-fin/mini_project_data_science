<h2>Merge Table</h2>

```python
# inner join 
new__ = pd.merge(df,lon_lat_df,on="District",how="inner")

```

```python
# Remove leading and trailing spaces and standardize to lower case
df['District'] = df['District'].str.strip().str.lower()
lon_lat_df['District'] = lon_lat_df['District'].str.strip().str.lower()
```

```python
# inner join 
merged_df = pd.merge(df,lon_lat_df,on='District',how='inner')
```

```python
#check the shape
merged_df.shape
```

```python
#check the difference 
difference_district = df[~df['District'].isin(lon_lat_df['District'])]
```

```python
#select only district column and only unique value
difference_district=difference_district.loc[:,'District'].unique()
```

```python
difference_district_lon_lat = lon_lat_df[~lon_lat_df['District'].isin(df['District'])]
```

```python
# show the difference district 
difference_district_lon_lat=difference_district_lon_lat['District'].unique()
```

```python
#sort the array
difference_district.sort() 
difference_district_lon_lat.sort()
```

```python
print(difference_district)
print(difference_district_lon_lat)
```

```python
#replace the district name 
for index in range(len(difference_district)):
    print(f"Replacing district name :{difference_district[index]}  to {difference_district_lon_lat[index]} ")
    df['District']=df['District'].str.replace(difference_district[index],difference_district_lon_lat[index])
```

```python
#now merge the datasets 
merged_df = pd.merge(df,lon_lat_df,on='District',how='inner')
```

```python
merged_df.shape
```

```python
#reoder the columns
new_columns_order = ['District', 'Zone', 'Development Region',  'Geographical Region', 'Latitude', 'Longitude','School Number', 'School_code', 'School_name',
       'ENGLISH',
       'NEPALI', 'MATHS', 'SCIENCE', 'SOCIAL STUDIES',
       'HEALTH POPULATION AND ENVIRONMENT', 'DISTINCTION',
       'FIRST DIVISION', 'SECOND DIVISION', 'THIRD DIVISION', 'PASS', 'FAIL',
       'PASS PERCENT']
merged_df= merged_df[new_columns_order]
#drop the  School Number
merged_df=merged_df.drop(columns=['School Number'])
```

```python
#reset index of  the dataset 
merged_df.reset_index(inplace=True)
```

```python
#save the merge dataset 
merged_df.to_csv("dataset/merged_clean_data.csv",index=False)
```
